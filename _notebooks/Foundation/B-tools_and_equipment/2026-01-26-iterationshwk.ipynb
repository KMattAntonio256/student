```python
fruits = ["Heart Shaped Herb", "Yami Yami no Mi", "Gomu Gomu no Mi"]

# Loop Type 1: For Loop
print("=== Using For Loop ===")
for i in range(len(fruits)):
    print(f"{i + 1}. {fruits[i]}")

# Loop Type 2: While Loop (different condition - counts down)
print("\n=== Using While Loop ===")
index = len(fruits) - 1
while index >= 0:
    print(f"Reverse order #{len(fruits) - index}: {fruits[index]}")
    index -= 1
```
