```python
let fruits = ["Heart Shaped Herb", "Yami Yami no Mi", "Gomu Gomu no Mi"];

// Loop Type 1: For Loop
console.log("=== Using For Loop ===");
for (let i = 0; i < fruits.length; i++) {
    console.log(`${i + 1}. ${fruits[i]}`);
}

// Loop Type 2: While Loop (different condition - counts down)
console.log("\n=== Using While Loop ===");
let index = fruits.length - 1;
while (index >= 0) {
    console.log(`Reverse order #${fruits.length - index}: ${fruits[index]}`);
    index--;
}
```
